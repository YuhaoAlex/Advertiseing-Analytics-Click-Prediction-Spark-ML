```
#read data from dataset
trainDataPath = "FileStore/tables/filtered_train.csv"
df = spark.read.format('csv')\
                .option('header','true')\
                .option('inferSchema', 'true')\
                .load(trainDataPath)
display(df)
```

```
df.printSchema()
```

```
df.count()
```

```
#decrease the RDD data partitions to 4
#store data as parquet format, 4 parts
df.coalesce(4)\
  .write\
  .mode("overwrite")\
  .parquet("/project2/train.csv")
```
